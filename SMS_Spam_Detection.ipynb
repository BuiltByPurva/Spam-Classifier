## Multicollinearity
Multicollinearity occurs when two or more independent (input) variables in a regression model are highly correlated with each other.
E.g num_of_words and num_of_sentences are highly correlated to character_len 

So, when we build model we will keep only one feature because it can create confusion for model. 
Thus, variable that is highly correlated to target variable, in our case is character_len

## 3. Data Preprocessing
1. Lower Case
2. Tokenization - break into words 
3. Removing special characters
4. Removing stop words and punctuation - words which are used for sentence formation but actually doesn't have much importance in the sentence e.g 'is', 'the', 'of'. 
5. Stemming/Lemmization - Convert to base words/Root form of words e.g dancing and dances are converted to dance

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords.words('english')

import string 
string.punctuation

from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
ps.stem('loving')
```

```python
from nltk.corpus import stopwords

def transform_text(text):
    text=text.lower()
    text=nltk.word_tokenize(text)
    y=[]
    for i in text:
        if i.isalnum():
            y.append(i)
            
    text=y[:]
    y.clear()
    for i in text:
        if i not in stopwords.words('english') and i not in string.punctuation:
            y.append(i)
            
    text=y[:]
    y.clear()
    for i in text:
        y.append(ps.stem(i))
        
    return " ".join(y)
transform_text('I loved the youtube lecture on machine learning')
```

```python
df['transformed_text']=df['text'].apply(transform_text)
```

## 4. Model Building 

Since, there is textual data, we are going to use *Navie Bayes* model.

Naive Bayes is a probabilistic machine learning algorithm based on Bayes' Theorem with a “naive” assumption that features are independent of each other.

It’s very popular for:
1. Spam detection 
2. Sentiment analysis 
3. Document classification 

A ML model needs numerical input, our target variable is numerical but text isn't numerical, so we need to convert them into vector. 
Following are methods used to convert text to vectors:
1. Bag of Words - take count of frequent words and then check them how many times it is repeated in the sentence.
2. TF-IDF
3. Word to Vect 

## Vectorizing Data by using Bag Of Words and TF-IDF

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
cv= CountVectorizer()
tfidf=TfidfVectorizer(max_features=3000)
```

```python
X= tfidf.fit_transform(df['transformed_text']).toarray()
X
```

```python
X.shape
```

```python
Y= df['target'].values
Y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score
```

```python
X_train, X_test, Y_train, Y_test=train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB 
gnb= GaussianNB()
mnb= MultinomialNB()
bnb= BernoulliNB()
```

```python
gnb.fit(X_train, Y_train)
Y_pred1=gnb.predict(X_test)
print(accuracy_score(Y_test, Y_pred1))
print(confusion_matrix(Y_test, Y_pred1))
print(precision_score(Y_test, Y_pred1))
```

```python
mnb.fit(X_train, Y_train)
Y_pred2=mnb.predict(X_test)
print(accuracy_score(Y_test, Y_pred2))
print(confusion_matrix(Y_test, Y_pred2))
print(precision_score(Y_test, Y_pred2))
```

```python
bnb.fit(X_train, Y_train)
Y_pred3=bnb.predict(X_test)
print(accuracy_score(Y_test, Y_pred3))
print(confusion_matrix(Y_test, Y_pred3))
print(precision_score(Y_test, Y_pred3))
```

Since, data is imbalanced we check for precision score and it is highest is TF-IDF MultinomialNB as comapred to CountVectorizer of BernoulliNB.
Thus, we use MultinomialNB of TfidfVectorizer
